### Part-I: Write a method for sigmoid function

```python
#1. Write a function, mysigmoid(x), that takes the real number x and returns the sigmoid value using math.exp().

import math
def mysigmoid(x):
    sig = 1/(1+math.exp(-x))
    return sig
```

```python
#2. Call mysigmoid() with x=4 and print the sigmoid value of 4.

mysigmoid(4)
```

```python
#3. Now, find the sigmoid value for x=[1,2,3]. Observe the results.
 
x = [1,2,3]
for i in x:
    print(mysigmoid(i))
```

```python
#4. Rewrite mysigmoid() using np.exp() function.

import math
import numpy as np
def mysigmoidnp(x):
    sig = 1/(1+np.exp(-x))
    return sig
```

```python
#5. Now call your function with x=[1,2,3] and observe the results.

x = [1,2,3]
for i in x:
    print(mysigmoidnp(i))
```

### Part-II Gradient or derivative of sigmoid function

```python
def sig_derivative(x):
    s = mysigmoid(x)
    s_derivative = s*(1-s)
    return s_derivative
```

```python
x = [1,2,3]
for i in x:
    print(sig_derivative(i))
```

### Part-IV: Write a method normalizeRows()

```python
def normalizeRows(x):
    x_norm = np.linalg.norm(x, ord=2,axis=1,keepdims=True)
    x = x/x_norm
    return x
```

```python
x = np.array([[0,3,4],[1,6,4]])
print('normalizeRows(x) = ' + str(normalizeRows(x)))
```

### Part-VI : Implement L1 and L2 loss functions

```python
def L1(yhat, y):
    """
    Arguments:
    yhat -- vector of size m (predicted labels)
    y -- vector of size m (true labels)
    
    Returns:
    loss -- the value of the L1 loss function defined above
    """
    
    ### START CODE HERE ### ( 1 line of code)
    loss = np.sum(abs(y - yhat), axis = 0, keepdims = True)
    ### END CODE HERE ###
    
    return loss
yhat = np.array([.9, 0.2, 0.1, .4, .9])
y = np.array([1, 0, 0, 1, 1])
print("L1 = " + str(L1(yhat,y)))
```

```python
def L2(yhat, y):
    """
    Arguments:
    yhat -- vector of size m (predicted labels)
    y -- vector of size m (true labels)
    
    Returns:
    loss -- the value of the L2 loss function defined above
    """
    
    ### START CODE HERE ### ( 1 line of code)
    loss = np.dot((y - yhat),(y - yhat))
    ### END CODE HERE ###
    
    return loss
yhat = np.array([.9, 0.2, 0.1, .4, .9])
y = np.array([1, 0, 0, 1, 1])
print("L2 = " + str(L2(yhat,y)))
```
